## **3. Data Wrangling**

In this section, we prepare data for analysis and modeling. This covers merging all the datasets into a single dataset, fixing data types. handling missing values, and creating useful features.After this phase the dataset is clean, consistent, and ready for further analysis.

**Convert Date column to Datetime**

```python
# Converting 'Date' column with correct format (DD-MM-YYYY) to Datetime format

sales_df['Date'] = pd.to_datetime(sales_df['Date'], dayfirst=True)
features_df['Date'] = pd.to_datetime(features_df['Date'], dayfirst=True)

print("Date columns converted successfully!")
```

**Merge Datasets**

```python
# Merging sales and features dataset

merged_df = pd.merge(sales_df, features_df, on=['Store', 'Date'], how='left')

# Merging with stores dataset

merged_df = pd.merge(merged_df, stores_df, on='Store', how='left')

print("Datasets merged successfully!")
print("Merged Dataset Shape:", merged_df.shape)
```

**Check Missing Values After Merge**

```python
# Checking missing values

print("Missing values after merging")
print(merged_df.isnull().sum())
```

**Handle Essential Missing Values**

```python
# Filling MarkDown missing values with 0 (no promotion assumption)

markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']

for col in markdown_cols:
    merged_df[col] = merged_df[col].fillna(0)

print("MarkDown missing values handled successfully!")
```

**Handle Duplicate Holiday Columns**

```python
# We have IsHoliday_x and IsHoliday_y after merge
# Keeping one and dropping the other

merged_df.rename(columns={'IsHoliday_x': 'IsHoliday'}, inplace=True)
merged_df.drop('IsHoliday_y', axis=1, inplace=True)

print("Duplicate IsHoliday columns handled!")
```

## **7. Hypothesis Testing**

This section focuses on validating insights derived from exploratory data analysis using statistical methods. Hypothesis testing helps determine whether observed patterns in the data are statistically significant or occurred by chance. By formulating hypotheses and applying appropriate statistical tests, we aim to draw reliable conclusions that support data-driven decision-making and strengthen the overall credibility of the analysis.


#### **Hypothetical Statement - 1**

*"Sales are higher during holidays compared to non-holidays"*

**Step 1: Define Hypotheses**
> Null Hypothesis (H₀):
> - There is no significant difference in weekly sales between holiday and non-holiday weeks.
>
> Alternative Hypothesis (H₁):
> - Weekly sales during holidays are significantly higher than non-holiday weeks.

**Step 2: Statistical Test**

> *Independent T-Test*
>
> **Why?**
>
> We are comparing the mean sales of two independent groups (holiday vs non-holiday).

```python
# Importing required library
from scipy.stats import ttest_ind

# Separating data
holiday_sales = merged_df[merged_df['IsHoliday'] == True]['Weekly_Sales']
non_holiday_sales = merged_df[merged_df['IsHoliday'] == False]['Weekly_Sales']

# Performing Independent T-Test
t_stat, p_value = ttest_ind(holiday_sales, non_holiday_sales, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

**Interpretation**

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis → Sales during holidays are significantly higher.")
else:
    print("Fail to Reject Null Hypothesis → No significant variation")
```

#### **Hypothetical Statement - 2**

*“Store Type has an impact on Weekly Sales”*

**Step 1: Define Hypotheses**
> Null Hypothesis (H₀):
> - Store type has no impact on weekly sales.
>
> Alternative Hypothesis (H₁):
> - Store type significantly affects weekly sales.

**Step 2: Statistical Test**

> *One-Way ANOVA Test*
>
> **Why?**
>
> We are comparing means of more than two groups (Type A, B, C).

```python
# Importing ANOVA test
from scipy.stats import f_oneway

# Grouping data by store type
sales_A = merged_df[merged_df['Type'] == 'A']['Weekly_Sales']
sales_B = merged_df[merged_df['Type'] == 'B']['Weekly_Sales']
sales_C = merged_df[merged_df['Type'] == 'C']['Weekly_Sales']

# Performing ANOVA test
f_stat, p_value = f_oneway(sales_A, sales_B, sales_C)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

**Decision Rule**

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis → Store type has a significant impact on sales.")
else:
    print("Fail to Reject Null Hypothesis → No significant variation")
```

#### **Hypothetical Statement - 3**

*“MarkDown1 (Promotions) affects Weekly Sales”*

**Step 1: Define Hypotheses**
> Null Hypothesis (H₀):
> - There is no relationship between MarkDown1 and Weekly Sales.
>
> Alternative Hypothesis (H₁):
> - There is a significant relationship between MarkDown1 and Weekly Sales.

**Step 2: Statistical Test**
> *Pearson Correlation*
>
> **Why?**
>
> To measure linear relationship between two continuous variables.

```python
# Importing correlation test
from scipy.stats import pearsonr

# Performing Pearson Correlation
corr, p_value = pearsonr(merged_df['MarkDown1'], merged_df['Weekly_Sales'])

print("Correlation Coefficient:", corr)
print("P-Value:", p_value)
```

**Decision Rule**

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis → Promotions (MarkDown1) significantly affect sales.")
else:
    print("Fail to Reject Null Hypothesis → No relationship")
```

#### **Time-Series Model : Prophet**

---

Model: **Prophet**

Prophet is a time-series forecasting model developed by Facebook.

**Why this model?**

1. Handles seasonality and trend automatically.
2. Works well with business data (sales, demand).
3. Handles missing values and outliers effectively.
4. Easy to interpret and implement.

**Objective:**

---

**Prepare Data**

Prophet requires:

- ds → Date
- y → Target

```python
# Reset index
df_prophet = df_ts_final.reset_index()

# Aggregate sales by date 
df_prophet = df_prophet.groupby('Date')['Weekly_Sales'].sum().reset_index()

# Rename columns for Prophet
df_prophet.columns = ['ds', 'y']

print("Data shape after aggregation:", df_prophet.shape)
df_prophet.head()
```

**Train-Test Split (Time-Based)**

```python
# Split 80-20

split_index = int(len(df_prophet) * 0.8)

train_prophet = df_prophet.iloc[:split_index]
test_prophet = df_prophet.iloc[split_index:]

print("Train:", train_prophet.shape)
print("Test:", test_prophet.shape)
```

**Installing Prophet**

```python
# Install if needed
%pip install prophet
```

**Model Training**

```python
from prophet import Prophet

# Initialize model
prophet_model = Prophet()

# Train
prophet_model.fit(train_prophet)

print("Prophet model trained successfully!")
```

**Forecast**

```python
# Create future dataframe

future = prophet_model.make_future_dataframe(
    periods=len(test_prophet), 
    freq='W'
)

# Forecast
forecast = prophet_model.predict(future)
```

**Extract Predictions**

```python
# Extract only test predictions

forecast_test = forecast.tail(len(test_prophet))

y_pred_prophet = forecast_test['yhat'].values
y_true_prophet = test_prophet['y'].values
```

**Evaluation Metrics**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

rmse_prophet = np.sqrt(mean_squared_error(y_true_prophet, y_pred_prophet))
mae_prophet = mean_absolute_error(y_true_prophet, y_pred_prophet)
r2_prophet = r2_score(y_true_prophet, y_pred_prophet)

print("RMSE:", rmse_prophet)
print("MAE:", mae_prophet)
print("R2 Score:", r2_prophet)
```

**Forecast vs Actual Plot**

```python
plt.figure(figsize=(10,5))

plt.plot(test_prophet['ds'], y_true_prophet, label="Actual")
plt.plot(test_prophet['ds'], y_pred_prophet, label="Forecast")

plt.title("Prophet Forecast vs Actual")
plt.legend()
plt.show()
```

**Hyperparameter Tuning**

```python
# Tuned model

prophet_model_tuned = Prophet(
    changepoint_prior_scale=0.5,
    seasonality_prior_scale=10
)

# Fitting the tuned model
prophet_model_tuned.fit(train_prophet)

# Forecasting tuned model
forecast_tuned = prophet_model_tuned.predict(future)

forecast_test_tuned = forecast_tuned.tail(len(test_prophet))

# Prediction
y_pred_prophet_tuned = forecast_test_tuned['yhat'].values

# Tuned models metrics
rmse_prophet_tuned = np.sqrt(mean_squared_error(y_true_prophet, y_pred_prophet_tuned))
mae_prophet_tuned = mean_absolute_error(y_true_prophet, y_pred_prophet_tuned)
r2_prophet_tuned = r2_score(y_true_prophet, y_pred_prophet_tuned)

print("Tuned RMSE:", rmse_prophet_tuned)
print("Tuned MAE:", mae_prophet_tuned)
print("Tuned R2:", r2_prophet_tuned)
```

#### **Clustering Model: K-Means**

---

Model: **K-Means Clustering**

K-Means is an unsupervised learning algorithm used to group similar data points
into clusters.

**Why this model?**

1. Helps segment stores/departments based on sales patterns.
2. Useful for identifying high-performing and low-performing groups.
3. Enables targeted business strategies.

**Objective:**

To segment stores/departments into meaningful groups based on sales behavior.

---

**Prepare Data for Clustering**

```python
# Aggregate data by Store

df_cluster = df_ts_final.reset_index()

df_cluster = df_cluster.groupby('Store').agg({
    'Weekly_Sales': 'mean',
    'MarkDown1': 'mean',
    'MarkDown2': 'mean',
    'MarkDown3': 'mean',
    'MarkDown4': 'mean',
    'MarkDown5': 'mean',
    'CPI': 'mean',
    'Unemployment': 'mean',
    'Size': 'first'
}).reset_index()

df_cluster.head()
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_cluster = df_cluster.drop('Store', axis=1)

X_scaled = scaler.fit_transform(X_cluster)
```

**Find Optimal Clusters (Elbow Method)**

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(6,4))
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

**Apply K-Means**

```python
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)

df_cluster['Cluster'] = kmeans_model.fit_predict(X_scaled)

df_cluster.head()
```

**Visualization**

```python
plt.figure(figsize=(8,5))

plt.scatter(df_cluster['Weekly_Sales'], df_cluster['Size'],
            c=df_cluster['Cluster'], cmap='viridis')

plt.xlabel("Average Sales")
plt.ylabel("Store Size")
plt.title("Store Segmentation")
plt.show()
```

**Cluster Summary**

```python
cluster_summary = df_cluster.groupby('Cluster').mean()

print(cluster_summary)
```

**Evaluation (Silhouette Score)**

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, df_cluster['Cluster'])

print("Silhouette Score:", score)
```

**Hyperparameter Tuning**

```python
# Try different K values

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    
    print(f"K={k}, Silhouette Score={score}")
```

---

Model: **K-Means Clustering**

**How we selected clusters:**

- Elbow graph suggested 2 or 3 clusters
